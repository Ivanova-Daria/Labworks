```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
df=pd.read_csv('aug_train.csv', sep=',')
```

# **Box plots**

```python
sns.boxplot(df['training_hours'])
```

```python
sns.kdeplot(df['training_hours'])
```

```python
df.head()
```

# **Violin plot**

```python
sns.violinplot(x='gender',y='training_hours', data =df)
```

# **Count plots**

```python
sns.countplot(df['education_level'])
```

# **Heatmap**

```python
sns.heatmap(df.corr(),annot=True,cmap='YlGnBu')
```

# **Facet grid**

```python
g=sns.FacetGrid(df, col="target", row='gender')
g.map(plt.hist, 'training_hours')
```

# **Regplot**

```python
df1 = df.copy(deep=True)

df1['experience']=df1['experience'].str.replace(r">","").str.replace(r"<","")
df1['experience']=df1['experience'].astype('float64')
df1['experience']=df1['experience'].fillna(0)
```

```python
sns.regplot(x='training_hours', y='experience',data=df1)
```

# **lmplot**

```python
sns.lmplot(x='training_hours', y='experience',row='target', data = df1, col='gender')
```

# **Swarm plot**

```python
sns.swarmplot(x='experience',y='training_hours',data=df1)
plt.show()
```

```python
sns.pairplot(data=df1)
```

```python
sns.jointplot(x='experience',y='training_hours',data=df1)
```
