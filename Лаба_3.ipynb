```
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```
def af(y, x):
  return x-y
```

```
x = np.linspace(-1,1,100)
y0 = 1
y = odeint(af ,y0 , x)
y = np.array(y).flatten()
```

```
fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_xlabel('$x$')
ax.set_ylabel('$y$')
ax.plot(x,y)
```
